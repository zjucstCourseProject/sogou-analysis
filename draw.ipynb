```python
from pyecharts import Pie, Bar
```

```python
def get_map(raw_data):
    lines = raw_data.split("\n")

    data_map = {}
    for line in lines:
        key, word = line.split("\t")
        data_map[key] = float(word)
    return data_map
```

```python
raw_data = '''百度	38437
baidu	18312
人体艺术	14474
4399小游戏	11438
qq空间	10316
优酷	10158
新亮剑	9654
馆陶县县长闫宁的父亲	9127
公安卖萌	8192
百度一下 你就知道	7504
百度一下	7104
4399	7041
魏特琳	6665
qq网名	6148
7k7k小游戏	5985
黑狐	5610
儿子与母亲不正当关系	5496
新浪微博	5369
李宇春体	5310
新疆暴徒被击毙图片	4997
hao123	4834
123	4829
4399洛克王国	4112
qq头像	4085
nba	4027
龙门飞甲	3917
qq个性签名	3880
张去死	3847
cf官网	3729
凰图腾	3632'''

top30_map = get_map(raw_data)
print(top30_map)
```

```python
# 绘制top30的柱状图
bar = Bar(title="2011年12月30日频度前30的搜索词")
columns = list(top30_map.keys())
data = list(top30_map.values())
bar.add("搜索量", columns, data, mark_line=["average"], mark_point=["max", "min"])
bar.render()
bar
```

```python
pie = Pie("百度相关搜索占比")
columns = ["百度相关"]
data = [0]
key_baidu_rela = ["baidu", "百度", "百度一下 你就知道", "百度一下", "hao123"]
for key in key_baidu_rela:
    if key in key_baidu_rela:
        data[0] += top30_map[key]
for key in top30_map.keys():
    if key not in key_baidu_rela:
        columns.append(key)
        data.append(top30_map[key])
pie = Pie("搜索占比")
pie.add("搜索量", columns, data, is_legend_show=False, is_label_show=True)
pie
```

```python
raw_data =  '''http://www.baidu.com/	0.014746762770364151
http://www.4399.com/	0.00380309355610148
http://www.hao123.com/	0.0028676705446953996
http://www.youku.com/	0.002817069299904778
http://qzone.qq.com/	0.002584063567963772
http://www.7k7k.com/	0.0016652409649277372
http://weibo.com/	0.001509437132153451
http://cf.qq.com/	0.001508837117393088
http://www.xixiwg.com/	0.0014086346524124493
http://www.12306.cn/	0.0013922342489625244'''
top10_map = get_map(raw_data)
# 改变下精度
for key in top10_map.keys():
    top10_map[key] = round(top10_map[key], 5)
print(top10_map)
```

```python
# 绘制top10的柱状图
pie = Pie(title="2011年12月30日点击频度前10的网址")
columns = list(top10_map.keys())
data = list(top10_map.values())
pie.add("点击量", columns, data, is_legend_show=False, is_label_show=True)
pie.render()
pie
```

```python
# 算上“其他”的点击占比图
pie = Pie(title="2011年12月30日网址点击频度占比")
columns.append("其他")
sum = 0
for value in data:
    sum += value
data.append(1 - sum)
pie.add("点击量", columns, data, is_legend_show=False, is_label_show=True, center=[50, 60])
pie.render()
pie
```

```python
# 游戏相关搜索占比
game_rela = ["http://www.4399.com/", "http://www.7k7k.com/", "http://cf.qq.com/", "http://www.xixiwg.com/"]
# 绘制top10的柱状图
pie = Pie(title="2011年12月30日游戏网址点击占比")
columns = ["游戏相关"]
data = [0]
for key in top10_map.keys():
    if key in game_rela:
        data[0] += top10_map[key]
    else:
        columns.append(key)
        data.append(top10_map[key])
pie.add("点击量", columns, data, is_legend_show=False, is_label_show=True)
pie.render()
pie
```
